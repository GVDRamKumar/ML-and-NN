
# ***CNN, RNN AND LSTM TUTORIAL***
This code demonstrates how to implement three different types of neural networks: Convolutional Neural Networks (CNN), Recurrent Neural Networks (RNN), and Long Short-Term Memory (LSTM) networks, using TensorFlow and Keras. These three models are applied to two well-known datasets:

MNIST Dataset: A dataset of handwritten digits (0-9) commonly used for image classification tasks.

IMDB Dataset: A dataset of movie reviews used for sentiment analysis (classifying reviews as positive or negative).

The goal is to show how each of these neural network architectures can be used for different types of data and problems.

# **Steps Covered in the Code:**
1. Data Loading and Preprocessing:

The MNIST dataset is loaded using tensorflow.keras.datasets.mnist, and the images are reshaped for each model (CNN, RNN, and LSTM). For CNN, the images are reshaped to (28, 28, 1) to match the expected input shape for convolutional layers. For RNN and LSTM, the images are reshaped to a sequence of 28 rows, where each row is a sequence of pixels.

2. Model Building:

For each model (CNN, RNN, and LSTM), a Sequential model is created using Keras.

The CNN model consists of convolutional layers for feature extraction, followed by max-pooling layers and dense layers for classification.

The RNN model consists of an embedding layer followed by a SimpleRNN layer for processing the sequential data, and a final dense layer for classification.

The LSTM model uses the LSTM layer to capture long-term dependencies in the input sequences, followed by a dense output layer for classification.

3. Model Compilation:

All models are compiled using the Adam optimizer and categorical crossentropy loss (for multi-class classification). The metric used to evaluate the models is accuracy.

4. Model Training:

Each model is trained using the fit() method, with the training data (x_train) and the labels (y_train). The models are trained for 5 epochs, and the performance is validated using the test set (x_test, y_test).

5. Model Evaluation:

After training, the models are evaluated on the test data to see how well they generalize to unseen data. The test accuracy is printed for each model.

6. Visualization:

Finally, the accuracy of all three models is plotted over the epochs to compare their performance during training.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, SimpleRNN, LSTM, Embedding
from tensorflow.keras.datasets import mnist, imdb
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 1. CNN Model for MNIST Dataset
# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape data to be (28, 28, 1) because CNN expects 3D input (height, width, channels)
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0  # Normalize to 0-1
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0
y_train = to_categorical(y_train, 10)  # One-hot encoding labels
y_test = to_categorical(y_test, 10)

# Initialize the CNN model
cnn_model = Sequential()

# Add a convolutional layer with 32 filters, 3x3 kernel, and ReLU activation
cnn_model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))

# Add a max-pooling layer to reduce the spatial dimensions
cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

# Another convolutional and max-pooling layer
cnn_model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten the 2D output to 1D so we can pass it to the dense layers
cnn_model.add(Flatten())

# Add a dense layer for classification
cnn_model.add(Dense(128, activation='relu'))

# Output layer with 10 neurons for 10 classes (digits 0-9)
cnn_model.add(Dense(10, activation='softmax'))

# Compile the CNN model with Adam optimizer and categorical crossentropy loss
cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the CNN model
cnn_history = cnn_model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))

# Evaluate the performance on the test set
cnn_test_loss, cnn_test_acc = cnn_model.evaluate(x_test, y_test)
print(f"CNN Test accuracy: {cnn_test_acc}")

# Show a sample test image
plt.imshow(x_test[0].reshape(28, 28), cmap='gray')
plt.title(f"Predicted Label: {np.argmax(cnn_model.predict(x_test[0].reshape(1, 28, 28, 1)))}")
plt.show()


# 2. RNN Model for IMDB Dataset
# Load IMDB dataset for sentiment analysis
(x_train_imdb, y_train_imdb), (x_test_imdb, y_test_imdb) = imdb.load_data(num_words=10000)

# Pad sequences so they all have the same length (500 words)
x_train_imdb = pad_sequences(x_train_imdb, maxlen=500)
x_test_imdb = pad_sequences(x_test_imdb, maxlen=500)

# Initialize the RNN model
rnn_model = Sequential()

# Add an embedding layer for converting words to dense vectors
rnn_model.add(Embedding(input_dim=10000, output_dim=128, input_length=500))

# Add an RNN layer (SimpleRNN) with 128 units
rnn_model.add(SimpleRNN(128, activation='tanh'))

# Output layer for binary classification (positive/negative review)
rnn_model.add(Dense(1, activation='sigmoid'))

# Compile the RNN model
rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the RNN model
rnn_history = rnn_model.fit(x_train_imdb, y_train_imdb, epochs=5, batch_size=64, validation_data=(x_test_imdb, y_test_imdb))

# Evaluate the RNN model on the test set
rnn_test_loss, rnn_test_acc = rnn_model.evaluate(x_test_imdb, y_test_imdb)
print(f"RNN Test accuracy: {rnn_test_acc}")


# 3. LSTM Model for MNIST Dataset
# Load the MNIST dataset
(x_train_lstm, y_train_lstm), (x_test_lstm, y_test_lstm) = mnist.load_data()

# Normalize the data
x_train_lstm = x_train_lstm.astype('float32') / 255.0
x_test_lstm = x_test_lstm.astype('float32') / 255.0

# Reshape data to fit LSTM input - (samples, timesteps, features)
# We want each pixel row to be a timestep, with 1 feature (pixel value)
x_train_lstm = x_train_lstm.reshape(x_train_lstm.shape[0], 28, 28)  # Reshape to (60000, 28, 28)
x_test_lstm = x_test_lstm.reshape(x_test_lstm.shape[0], 28, 28)    # Reshape to (10000, 28, 28)

# One-hot encode labels
y_train_lstm = to_categorical(y_train_lstm, 10)
y_test_lstm = to_categorical(y_test_lstm, 10)

# Initialize the LSTM model
lstm_model = Sequential()

# Add an LSTM layer with 50 units
lstm_model.add(LSTM(50, activation='relu', input_shape=(28, 28)))

# Add a dense output layer with 10 units for the 10 classes (digits)
lstm_model.add(Dense(10, activation='softmax'))

# Compile the LSTM model
lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the LSTM model
lstm_history = lstm_model.fit(x_train_lstm, y_train_lstm, epochs=5, batch_size=64, validation_data=(x_test_lstm, y_test_lstm))

# Evaluate the LSTM model
lstm_test_loss, lstm_test_acc = lstm_model.evaluate(x_test_lstm, y_test_lstm)
print(f"LSTM Test accuracy: {lstm_test_acc}")


# Visualizing the training accuracy of the three models (CNN, RNN, LSTM)
plt.plot(cnn_history.history['accuracy'], label='CNN')
plt.plot(rnn_history.history['accuracy'], label='RNN')
plt.plot(lstm_history.history['accuracy'], label='LSTM')
plt.title('Model Accuracy Comparison')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

automatically extract spatial features from images.
*   RNNs are ideal for sequential data, making them great for text-based tasks like sentiment analysis.


*   LSTMs improve on RNNs by handling long-term dependencies in sequences, making them more effective in complex sequential data problems.

Each model has its strengths and is suited to specific tasks, with CNNs being the best for images and RNNs/LSTMs for sequential data. This comparison highlights how these models can be applied to real-world problems in fields like computer vision and natural language processing.
